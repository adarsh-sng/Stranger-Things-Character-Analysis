## Transforming the data for our needs

```python

import pandas as pd
import re


df = pd.read_csv(path, encoding='Windows-1252')

def extract_words(series):
    return ' '.join(re.findall(r'\b\w+\b', ' '.join(series)))


character_dialogues = df.groupby('name')['line'].apply(extract_words).reset_index()

def count_words(text):
    words = re.findall(r'\b\w+\b', str(text))
    return len(words)

character_dialogues['Number of Words'] = character_dialogues['line'].apply(count_words)


result = character_dialogues.sort_values(by='Number of Words', ascending=False)

```

#Training the data

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
embeddings = cv.fit_transform(result['line']).toarray()
```

```python
embeddings.shape
```

```python
embeddings = embeddings.astype('float64')
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, verbose=1, random_state=123)
z = tsne.fit_transform(embeddings)
```

```python
z.shape
```

```python
result['x'] = z.T[0]
result['y'] = z.T[1]
#new_df['z'] = z.T[2]
```
